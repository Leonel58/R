```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Importamos los datos
df = pd.read_csv("/lineal.csv", sep=";", quotechar='"')

# Dividimos los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(df[["transacciones"]], df["ventas"], test_size=0.25)

# Entrenamos el modelo
model = linear_model.LinearRegression()
model.fit(X_train, y_train)

# Predecimos los valores de y para los datos de prueba
y_pred = model.predict(X_test)

# Calculamos el error
error = np.mean((y_test - y_pred)**2)

# Imprimimos el error
print("Error:", error)

# Graficamos los datos
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color="red")
plt.xlabel("Transacciones")
plt.ylabel("Ventas")
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Importamos los datos
df = pd.read_csv("/lineal.csv", sep=";", quotechar='"')

# Dividimos los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(df[["transacciones"]], df["ventas"], test_size=0.25)

# Entrenamos el modelo
model = linear_model.LinearRegression()
model.fit(X_train, y_train)

# Predecimos los valores de y para los datos de prueba
y_pred = model.predict(X_test)

# Calculamos el error
error = np.mean((y_test - y_pred)**2)

# Imprimimos la pendiente y el error
print("Pendiente (coeficiente):", model.coef_[0])
print("Error cuadrático medio:", error)

# Ecuación de la recta de regresión
slope = model.coef_[0]
intercept = model.intercept_
print(f"Ecuación de la recta de regresión: y = {slope:.2f}x + {intercept:.2f}")

# Predicción para un valor específico
valor_a_predecir = 300  # Cambia este valor según lo que quieras predecir
prediccion = model.predict([[valor_a_predecir]])
print(f"Predicción para {valor_a_predecir} transacciones: {prediccion[0]:.2f} ventas")

# Graficamos los datos
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color="red", label="Recta de Regresión")
plt.xlabel("Transacciones")
plt.ylabel("Ventas")

plt.legend()
plt.show()
```
