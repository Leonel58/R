##MODIFICACION DE VALORES FALTANTES

```python
BRAZIL_BK = BRAZIL.copy()
```

```python
BRAZIL_BK.head(1)
```

#### modificacion de valores faltantes con un valor fijo

```python
BRAZIL_BK = BRAZIL.copy()
BRAZIL_BK = BRAZIL_BK.fillna("3")
BRAZIL_BK.head(3)
```

```python
BRAZIL_BK = BRAZIL.copy()
BRAZIL_BK = BRAZIL_BK.fillna("2")
BRAZIL_BK.head(2)
```

#### modificacion de valores faltantes con un valor fijo en 1 columna

```python
BRAZIL_BK = BRAZIL.copy()
BRAZIL_BK = BRAZIL_BK.fillna("2")
BRAZIL_BK.head(2)
```

```python
BRAZIL_BK = BRAZIL.copy()
BRAZIL_BK["Stars"] = BRAZIL_BK["Stars"].fillna(BRAZIL_BK["Stars"].mean())
BRAZIL_BK.head(2)
```

```python
BRAZIL_BK["Reviews"] = BRAZIL_BK["Reviews"].fillna(BRAZIL_BK["Reviews"].mean())
BRAZIL_BK.head(2)
```

```python
BRAZIL_BK.head(1)
```

```python
BRAZIL_BK["Authors/Company"] = BRAZIL_BK["Authors/Company"].fillna(BRAZIL_BK["Authors/Company"].mean())
```

```python
BRAZIL_BK.head(2)
```

```python
BRAZIL_BK["Max_prices"] = BRAZIL_BK["Max_prices"].fillna(BRAZIL_BK["Max_prices"].quantile(0.25))
BRAZIL_BK.head(2)
```
